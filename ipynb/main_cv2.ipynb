```python
import cv2
import numpy as np
from numpy.lib.stride_tricks import as_strided
import os
import matplotlib.pyplot as plt
```

```python
dP = '../img'
sP = '../result'
```

```python
datalist = os.listdir(dP)
img_f = datalist[0]
image_org = cv2.imread(os.path.join(dP, img_f))
image = cv2.cvtColor(image_org, cv2.COLOR_BGR2GRAY)

fig = plt.figure(figsize=(20, 20))
plt.imshow(image, cmap='gray')
```

## cv2d

```python
img_cv = cv2.blur(image, (3, 3))

fig = plt.figure(figsize=(20, 20))
plt.imshow(img_cv, cmap='gray')
```

## adpt_thold

```python
img_adpt_thold = cv2.adaptiveThreshold(img_cv, 255, cv2.ADAPTIVE_THRESH_GAUSSIAN_C, cv2.THRESH_BINARY_INV, 25, 3)

fig = plt.figure(figsize=(20, 20))
plt.imshow(img_adpt_thold, cmap='gray')
```

## erosion

```python
k = np.ones((3, 3))
img_e = cv2.erode(img_adpt_thold, k, iterations = 1)
fig = plt.figure(figsize=(20, 20))
plt.imshow(img_e, cmap='gray')
```
